```python
import pandas as pd
df = pd.read_csv("data.csv") #my data is in my pole vault lift!!
df["成績"] = df["成績"].str.replace("M","")#the ob to int
df["成績"] = df["成績"].astype(float)
df["名次"] = df["名次"].astype(int)
df = df.drop("Unnamed: 0",axis=1)
df = df.drop("名稱",axis=1)
df = df.drop("組別",axis=1)
df = df.drop("風速",axis=1)
df = df.drop("Unnamed: 15",axis=1)
df = df.drop("備註",axis=1)
df = df.drop("比賽日期",axis = 1)
df = df.drop("全能運動",axis=1)
df = df.drop("代表單位",axis=1)
df = df.drop("層級",axis=1)
df.to_csv("newData.csv",encoding="utf-8")
```
